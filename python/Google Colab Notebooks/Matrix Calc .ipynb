```python
import numpy as np
```

```python
r1 = int(input("No. of Rows for m1: "))
c1=int(input("No. of Cols for m2: "))
ele=list(map(int,input().split()))

m1 = np.array(ele).reshape(r1,c1)

r2 = int(input("No. of Rows for m2: "))
c2=int(input("No. of Cols for m2: "))
ele=list(map(int,input().split()))

m2 = np.array(ele).reshape(r2,c2)
print(m1)
print(m2)
```

```python
33 # basic calc

opt;
while (opt<=9):

  opt=int(input("Enter Option: "))
  try:
    np.broadcast_shapes(m1.shape,m2.shape)

    if(opt ==1):
      print("add: ",m1+m2)
      continue
    elif(opt==2):
      print("subtract: ",m1-m2)
      continue
    elif(opt==3):
      print("multiply: ",m1*m2)
      continue
    elif(opt==4):
      print("divide: ",m1/m2)
      continue
    elif(opt==5):
      print(np.stack((m1,m2),axis=-1))
    else:
      print("exiting")
      break;
  except ValueError:
    print("Invalid")
```
