```python
data = pd.read_csv('data.csv')
print(data)
```

```python
numeric_data = data.select_dtypes(include='number',exclude='string')
print(numeric_data)
```

```python
matrix = numeric_data.to_numpy()
print(matrix)
```

```python
for x in matrix[0:5]:
  print("student {x} stats: ")
  row = x
  print("Total Marks: ", np.sum(row))
  print("Mean: ",np.mean(row))
  print("Median: ",np.median(row))
  print("Standard Deviation: ",np.std(row))
  print("Variance: ",np.var(row))
  print("Min: ",np.min(row))
  print("Max: ",np.max(row))
```
