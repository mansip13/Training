# loc vs iloc

## .loc (label-based indexing):
- Selects data based on the labels of rows and columns.
- When slicing, it is inclusive of the endpoint. For example, df.loc['start_label':'end_label'] includes both 'start\_label' and 'end\_label'.
- Can use boolean arrays for filtering based on conditions.
- Example: df.loc['row_label', 'column_label'] or df.loc[df['column'] > 5, ['col1', 'col2']].

##.iloc (integer-location based indexing):
- Selects data based on the integer positions of rows and columns (0-based indexing).
- When slicing, it is exclusive of the endpoint. For example, df.iloc[0:5]
- selects rows from index 0 up to (but not including) index 5.
- Does not support boolean arrays for filtering directly; requires integer or slice inputs.
- Example: df.iloc[0, 1] or df.iloc[0:5, 2:4].
